# Calling LLM Model

```python


def get_groq_llm():
    return ChatOpenAI(
        model = "openai/gpt-oss-120b",
        base_url = "https://api.groq.com/v1",
        api_key = os.getenv("Groq_API_KEY"),
        # model_kwargs = {"temperature": 0.7},
        max_tokens = 1000


    )
```

# Calling embeddings Model

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
def get_gemini_embeddings_model():
    return GoogleGenerativeAIEmbeddings(model= "models/gemini-embedding-001")
```

```python
embedding_function = get_gemini_embeddings_model()
```

```python
from langchain_community.vectorstores import Chroma
```

```python
retriver= db.as_retriever(search_type= "mmr", search_kwargs= {"k":3}) # mmr stands for maximal marginal relevance, it is a method used in information retrieval to select a subset of documents that are both relevant to the query and diverse from each other. 
                                                                    # The "k" parameter specifies the number of documents to retrieve.
```
